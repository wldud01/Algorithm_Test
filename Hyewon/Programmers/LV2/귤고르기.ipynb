<h2>귤고르기</h2>

```python
def solution(k, tangerine) :
    answer = 0
    a = {}
    for i in tangerine :
        if i in a :
            a[i] += 1
        else :
            a[i] = 1
    #tangerine에 담긴 각각 다른 종류의 사이즈별로 개수를 dict에 넣어줌
    
    a = dict(sorted(a.items(), key=lambda x: x[1], reverse=True))
    #dict에 저장된 값들을 key값이 아닌 value 기준 내림차순 정렬
    #ex) key:value일 때, value가 많은 순으로 정렬
    #(최소한의 종류를 출력하기 위해 많은 것들부터 상자에 담는다)
    
    for i in a :
        if k <=0 :
            return answer
        k-=a[i]
        answer+=1
    return answer
    #내림차순 정렬된 value값을 k(상자에 담을 수 있는 귤 개수)에서 빼주면서 반복한다.
    #answer 값에 1씩 카운트해주며(종류)
    #k가 0보다 작을 때(다 찼을 때) answer를 출력
```
