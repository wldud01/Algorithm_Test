<h2>멀리뛰기</h2>

```python
def solution(n) :
    if n<3 :
        return n
    d = [0]*(n+1)
    d[1] = 1
    d[2] = 2
    for i in range(3,n+1) :
        d[i]=d[i-1] + d[i-2]
    return d[n]%1234567
```

```python
solution(4)
```

계단을 1,2칸만 올라갈 수 있으므로 3칸을 오를 때는 1,2칸을 오르는 경우의 수를 합해주면되고, 점화식으로 나타내면 f(n) = f(n-1)+f(n-2) 피보나치 수열을 구할 때와 같다

```python
def jumpCase(num) :
    answer = 0
    
    if num ==1 :
        return 1
    elif num == 2 :
        return 2
    else :
        return jumpCase(num-1) + jumpCase(num-2)
```
