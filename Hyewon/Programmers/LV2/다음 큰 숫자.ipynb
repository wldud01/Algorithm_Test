<h3>문제접근</h3>

- 주어진 값에 대하여 이진법으로 변환 후 1의 개수 세는 함수 check 짜기
- n을 이진법으로 변환했을 때 1의 개수 세기
- n보다 큰 수 answer에 +1dmf 반복해가며 n의 이진법 변환 1의 개수와 같으면 break
- return answer

<h3>문제풀이</h3>

```python
def check(x) :
    binary = bin(x)
    one = binary.count('1')
    return one
```

```python
def solution(n) :
    answer = n
    num = check(n)
    while True :
        answer += 1
        if int(check(answer)) == num :
            return answer
```

```python
solution(n)
```

<h3>다른풀이</h3>

```python
def solution_a(n) :
    c = n+1
    while True :
        if bin(c).count('1') == bin(n).count('1') :
            return c
        c += 1
```

```python
solution_a(n)
```
